## QUES-18 With the help of hypothesis testing check whether fraudulent transactions of higher value than normal transactions?

```python
from scipy.stats import ttest_ind

# Assuming 'df' is your DataFrame containing transaction data

# Extract transaction amounts for fraudulent and non-fraudulent transactions
fraudulent_amounts = df[df['Class'] == 1]['Amount']
non_fraudulent_amounts = df[df['Class'] == 0]['Amount']

# Perform the t-test
t_statistic, p_value = ttest_ind(fraudulent_amounts, non_fraudulent_amounts, alternative='greater')

# Set the significance level
alpha = 0.05

# Check for significance
if p_value < alpha:
    print("Reject the null hypothesis: There is evidence that fraudulent transactions have higher values.")
else:
    print("Fail to reject the null hypothesis: There is no evidence that fraudulent transactions have higher values.")
```

# QUES-19) Perform ANOVA test for Statistical feature selection.

```python
import pandas as pd
import numpy as np

# Assuming 'df' is your DataFrame containing the dataset
# Replace 'Categorical' and 'Numeric' with your actual column names

# Generate a sample DataFrame (replace this with your actual data loading)
data = {
    'Categorical': np.random.choice(['A', 'B', 'C'], size=100),
    'Numeric': np.random.randint(1, 100, size=100)
}
df = pd.DataFrame(data)

# Extract unique categories from the categorical column
categories = df['Categorical'].unique()

# Calculate group means
group_means = [df[df['Categorical'] == category]['Numeric'].mean() for category in categories]

# Overall mean of the numerical feature
overall_mean = df['Numeric'].mean()

# Calculate between-group variability (SSB)
SSB = sum([(df[df['Categorical'] == category]['Numeric'].mean() - overall_mean) ** 2 for category in categories])

# Calculate within-group variability (SSW)
SSW = sum([sum((df[df['Categorical'] == category]['Numeric'] - group_means[i]) ** 2) for i, category in enumerate(categories)])

# Degrees of freedom
df_between = len(categories) - 1
df_within = len(df) - len(categories)

# Calculate F-statistic
F_statistic = (SSB / df_between) / (SSW / df_within)

# Calculate p-value using F-distribution
from scipy.stats import f

p_value = 1 - f.cdf(F_statistic, df_between, df_within)

# Set significance level
alpha = 0.05

# Check for significance
if p_value < alpha:
    print("Reject the null hypothesis: There are significant differences in means across the categories.")
else:
    print("Fail to reject the null hypothesis: There are no significant differences in means across the categories.")

```

## QUES-14. Treat this dataset as a binomial distribution where p is the probability that a person becomes a CEO above 50 years age. What is the probability that out of a random sample of 10 CEOs of Fortune 500 companies exactly 6 are above 50 years of age? 

```python
from scipy.stats import binom

# Define the dataset
ceo_df = pd.DataFrame({'Age': ages_of_ceos})

# Calculate the probability of success (p)
p = (ceo_df['Age'] > 50).mean()

# Define the parameters for the binomial distribution
n = 10  # number of trials (sample size)
k = 6   # number of successes

# Calculate the probability using the binomial distribution formula
probability_6_successes = binom.pmf(k, n, p)

print(f"The probability that exactly 6 out of 10 CEOs are above 50 years of age is approximately: {probability_6_successes:.4f}")
```

## QUES-15. A study claims that 25% of all Fortune 500 companies becoming a CEO are above 60 years of age. Using the Normal approximation of a Binomial distribution, find the probability that in a random sample of 300 Fortune 500 companies exactly 75 CEOs will be above 50 years of age. [ Note that the normal distribution can be used to approximate a binomial distribution if np>=5 and nq>=5 with the following correction for continuity P(X=z) = P(z-0.5 < X < z+0.5 ) ]from scipy.stats import norm.

```python
from scipy.stats import norm

# Given values
p = 0.25  # probability of success
n = 300   # number of trials (sample size)
k = 75    # number of successes

# Calculate np and npq
np_value = n * p
npq_value = n * p * (1 - p)

# Calculate the probability using the normal approximation
z_lower = (k - np_value) / (npq_value**0.5)
z_upper = (k + 0.5 - np_value) / (npq_value**0.5)

# Use the correction for continuity
probability_normal_approximation = norm.cdf(z_upper) - norm.cdf(z_lower)

print(f"The probability that exactly 75 CEOs out of 300 are above 50 years of age is approximately: {probability_normal_approximation:.4f}")
```

## Q16. Compute a 95% Confidence Interval for the true mean age of the population of CEOs for the given dataset using appropriate distribution.( State reasons as to why did you use a z or t distribution)

```python
import scipy.stats as stats
import numpy as np
from scipy.stats import t

# Sample statistics
sample_mean = np.mean(ages_of_ceos)
sample_std = np.std(ages_of_ceos, ddof=1)  # ddof=1 for sample standard deviation
sample_size = len(ages_of_ceos)

# Degrees of freedom for t-distribution
degrees_of_freedom = sample_size - 1

# Confidence level and alpha
confidence_level = 0.95
alpha = 1 - confidence_level

# Calculate the critical value from the t-distribution
t_critical = t.ppf(1 - alpha/2, degrees_of_freedom)

# Calculate the margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"The 95% confidence interval for the true mean age of CEOs is: {confidence_interval}")
```

## QUES-19 A researcher claims that currently 25% of all CEOS are above 60 years .Test his claim with an alpha =0.05 if out of a random sample of 30 CEOs only10 are above 60 years.

```python
from scipy.stats import norm

# Given data
sample_size = 30
observed_above_60 = 10
claimed_proportion = 0.25  # 25% claimed proportion
alpha = 0.05

# Calculate sample proportion
sample_proportion = observed_above_60 / sample_size

# Calculate standard error
standard_error = (claimed_proportion * (1 - claimed_proportion) / sample_size) ** 0.5

# Calculate Z-score
Z = (sample_proportion - claimed_proportion) / standard_error

# Calculate critical Z-values
critical_value_left = norm.ppf(alpha / 2)
critical_value_right = norm.ppf(1 - alpha / 2)

print(f"Z-score: {Z:.2f}")
print(f"Critical values: {critical_value_left:.2f} and {critical_value_right:.2f}")

# Check if Z-score falls in the critical region or calculate p-value
if Z < critical_value_left or Z > critical_value_right:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

## QUES-20

```python
"""20)
a. Hypotheses
1) Null Hypothesis (H0): The average CEO age in the UK is the same as the average CEO age in the USA.
Î¼ (UK)=Î¼ (USA)
2) Alternative Hypothesis (H1): The average CEO age in the UK is older than the average CEO age in the USA.

b. Significance Level
1) The significance level is the threshold for accepting or rejecting the null hypothesis. Common values are 0.05 or 0.01, denoted as Î±.
2) Let's use a significance level of Î±=0.05 for this test.

c. Statistical Test
Since we are comparing the means of two independent samples from different populations (UK and USA),
a two-sample t-test for independent samples will be appropriate."""

"""d. Test Results
1) To perform the t-test, we need sample means, sample standard deviations, and sample sizes for both the UK and USA.
Given the data provided:
UK sample size (n1) = 40
USA sample size (n2) = 40

Given s1 = 8.8 and s2 = 7.8
"""
import numpy as np
from scipy.stats import t

# Given data
UK_data = np.array([47, 49, 73, 50, 65, 70, 49, 47, 40, 43, 46, 35, 38, 40, 47, 39, 49, 37, 37, 36,
                    40, 37, 31, 48, 48, 45, 52, 38, 38, 36, 44, 40, 48, 45, 45, 36, 39, 44, 52, 47])
USA_data = np.array([47, 57, 52, 47, 48, 56, 56, 52, 50, 40, 46, 43, 44, 51, 36, 42, 49, 49, 40, 43,
                     39, 39, 22, 41, 45, 46, 39, 32, 36, 32, 32, 37, 33, 44, 49, 44, 44, 49, 32])

# Sample statistics
n1 = len(UK_data)
n2 = len(USA_data)
mean_UK = np.mean(UK_data)
mean_USA = np.mean(USA_data)
s1 = 8.8
s2 = 7.8

# Calculate t-test statistic
SE = np.sqrt((s1**2 / n1) + (s2**2 / n2))
t_statistic = (mean_UK - mean_USA) / SE

# Degrees of freedom
df = n1 + n2 - 2

# Calculate critical t-value for one-tailed test at alpha = 0.05
alpha = 0.05
critical_t_value = t.ppf(1 - alpha, df)

print(f"T-test statistic: {t_statistic:.2f}")
print(f"Critical t-value: {critical_t_value:.2f}")
```
